```python
import pandas as pd

# Assuming the DataFrames are already loaded as lichsudoidau, predata, and data

# Step 1: Merge lichsudoidau with predata on team1.id and home.id, team2.id and away.id
merged_df = pd.merge(lichsudoidau, pre_data, left_on=['team1.id', 'team2.id'], right_on=['home_id', 'away_id'])

# Step 2: Merge the result with data on home_name and away_name
final_df = pd.merge(merged_df, data, left_on=['home_name', 'away_name'], right_on=['home_name', 'away_name'])

# Step 3: Select the required columns to add to the final DataFrame
columns_to_add = [
    'team1.overall_win_ratio', 'team1.overall_draw_ratio', 'team1.overall_loss_ratio',
    'team1.h2h_win_ratio', 'team1.h2h_draw_ratio', 'team1.h2h_loss_ratio',
    'team2.overall_draw_ratio', 'team2.overall_loss_ratio',
    'team2.h2h_win_ratio', 'team2.h2h_draw_ratio', 'team2.h2h_loss_ratio'
]

# Ensure the columns exist in the final DataFrame
final_df = final_df[['home_name', 'away_name'] + columns_to_add]

# Display the final DataFrame
final_df
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
# Giả sử bạn đã có các DataFrame: final_dl và data

# Bước 1: Lấy các cột cần thêm từ final_dl ngoại trừ home_name và away_name
# Bước 2: Merge các cột này vào DataFrame data
merged_df = pd.merge(data, final_df, on=['home_name', 'away_name'], how='left', suffixes=('_x', '_y'))

# Lấy danh sách các cột cần giữ lại (không bao gồm các cột có hậu tố '_x' và '_y')
columns_to_keep = [col for col in merged_df.columns if not col.endswith('_x') and not col.endswith('_y')]

# Tạo DataFrame mới chỉ với các cột cần giữ lại
cleaned_df = merged_df[columns_to_keep]
cleaned_df
```
